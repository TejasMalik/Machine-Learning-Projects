## Matplotlib - pyplot

* Bar Graph
* Histogram
* Pie Chart
* Scatter or Bubble Plots
* Line Plot

```python
import matplotlib.pyplot as plt
```

## Bar Graph

```python
x = [30, 20, 50, 40]
y = [10, 30, 40, 70]
plt.bar(x,y, color = 'red', width = 3)
plt.title("Bar Graph")
plt.xlabel("x values")
plt.ylabel("y values")
plt.show()
```

```python
x1 = [1,3,5,7,9]
y1 = [23,45,67,89,100]
x2 = [2,4,6,8,10]
y2 = [12,45,89,95,54]
plt.title("Bar Graph")
plt.bar(x1,y1, color ="red")
plt.bar(x2,y2, color = "black")
plt.show()
```

## Scatter Plot

```python
x1 = [1,3,5,7,9]
y1 = [23,45,67,89,100]
x2 = [2,4,6,8,10]
y2 = [12,45,89,95,54]
plt.figure(figsize = (10,10))
plt.scatter(x1,y1, c ="red", s = 40, alpha = 0.7, marker='^')
plt.scatter(x2,y2, c = "black", s =50, alpha = 1, marker='s')
plt.show()
```

## Line Plot

```python
x1 = [1,3,5,7,9]
y1 = [23,45,67,89,100]
x2 = [2,4,6,8,10]
y2 = [12,45,89,95,54]
plt.plot(x1,y1, label = "line1")
plt.plot(x2,y2, label = "line2")
plt.legend(loc = 2)
plt.show()
```

## Pie Chart

```python
x = [23,45,67,89,100]
activities = ["A", "B", "C", "D", "E"]
cols = ["red", "blue", "pink", "yellow", "green"]
plt.figure(figsize=(15,8))
plt.pie(x, labels=activities, colors=cols, startangle=0, explode=(0,0,0.1,0,0), autopct='%1.2f%%')
plt.title('Pie Chart')
plt.legend(loc =4)
plt.show()
```

## NumPy

```python
import numpy as np
np.random.seed(1)
x = np.random.randint(0,101,20)
print(x)
```

## Histogram

```python
plt.hist(x,bins = 10,color ='magenta')
plt.show()
print(x)
```

## Mix Graphs

```python
x = [1,3,5,7,9]
y = [20,40,60,80,100]
plt.bar(x,y,label='bar')
plt.plot(x,y,label='plot',color ='red')
plt.scatter(x,y,label ='scatter',color ='black')
plt.title('bar-line')
plt.xlabel("x")
plt.ylabel('y')
plt.legend()
plt.show()
```

## Horizontal Graphs

```python
x = [1,3,5,7,9]
y = [20,40,60,80,100]
plt.barh(x,y,label='bar')
plt.title('horizontal bar')
plt.xlabel("y")
plt.ylabel('x')
plt.legend()
plt.show()
```
