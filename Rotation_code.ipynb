```python
filepath='D://Images for train//data'
dest='D://Images for train//Data_r'

def image_rotation_fixed_number_of_image(filepath,start_number,end_number,dest):
    for i in range(start_number,end_number+1):
        x = cv2.imread(filepath+'//'+'framei'+' '+'('+str(i)+')'+'.jpg')
        p=np.random.uniform(0,1,1)
        if p>0.1:
            if np.random.choice([1,2,3,4,5,6,7,8,9]) == 1 :
                img = imutils.rotate_bound(x, 90)
                cv2.imwrite(dest+'//'+'framei'+' '+'('+str(i)+')'+'.jpg',img)
            else :
                img =imutils.rotate_bound(x, 90)
                cv2.imwrite(dest+'//'+'framei'+' '+'('+str(i)+')'+'.jpg',img)
```

```python
image_rotation_fixed_number_of_image(filepath,1,100,dest)
```

```python
def image_rotation_all_images(filepath,dest):
    image_list = glob.glob(filepath+"/*.jpg")
    for i in image_list:
        x = Image.open(i)
        image_number = i.split("/")[1]
        p=np.random.uniform(0,1,1)
        if p>0.1:
            if np.random.choice([1,2,3,4,5,6,7,8,9]) == 1 :
                img = x.rotate(90)
                img.save(dest+"/"+image_number)
            else :
                img = x.rotate(270)
                img.save(dest+"/"+image_number)
```

```python
image_rotation_all_images("Oats_and_O_s","Rotated_Oats_and_O_s")
```

# Distributing data to train and test

```python
import glob
from shutil import copyfile
annotated_images = glob.glob("Rotated_PuffedRice/*.xml")
for i in annotated_images:
    x = i.split("/")[1]
    y= x.split(".")[0]
    p=np.random.uniform(0,1,1)
    if p < 0.05 :
        copyfile(i, "images/test/"+y+".xml")
        copyfile(i.split("/")[0]+"/"+y+".jpg","images/test/"+y+".jpg")
    else :  
        copyfile(i, "images/train/"+y+".xml")
        copyfile(i.split("/")[0]+"/"+y+".jpg","images/train/"+y+".jpg")
```
