```python
notes = []

for file in glob.glob("midi_songs/*.mid"):
    midi = converter.parse(file)
    
    print("parsing %s" %file)
    
    elements_to_parse = midi.flat.notes
    
    for ele in elements_to_parse:
        # If element is note, take its pitch value
        if isinstance(ele, note.Note):
            notes.append(str(ele.pitch))
    
    
        # If element is chord, split each note of chord then join them '+'
        elif isinstance(ele, chord.Chord):
            notes.append("+".join(str(n) for n in ele.normalOrder))
```

```python
with open("notes", "wb") as filepath:
    pickle.dump(notes, filepath)
```

```python
with open("notes", "rb") as f:
    notes = pickle.load(f)
```

```python
n_vocab = len(set(notes))
```

```python
print("Total Notes - ", len(notes))
print("Unique Notes - ", n_vocab)
```

## Preparing Sequential Data for LSTM

```python
# How many elements LSTM input should consider
sequence_length = 100
```

```python
pitchnames = sorted(set(notes))
```

```python
# Mapping between elements to Integer value
ele_to_int = dict((ele, num) for num, ele in enumerate(pitchnames))
```

```python
print(ele_to_int)
```

```python
network_input = []
network_output = []
```

```python
for i in range(len(notes) - sequence_length):
    seq_in = notes[i : i+sequence_length] # Contains 100 values
    seq_out = notes[i + sequence_length]

    network_input.append([ele_to_int[ch] for ch in seq_in])
    network_output.append(ele_to_int[seq_out])
```

```python
# No. of examples
n_patterns = len(network_input)
n_patterns
```

```python
network_input = np.reshape(network_input, (n_patterns, sequence_length, 1))
print(network_input.shape)
```

```python
normalised_network_input = network_input/float(n_vocab)
```

```python
# Network Output are classes and convert them into one hot vector
network_output = np_utils.to_categorical(network_output)
```

```python
network_output.shape
```

## Create a Model

```python
from keras.models import Sequential, load_model
from keras.layers import *
from keras.callbacks import ModelCheckpoint, EarlyStopping
```

```python
model = Sequential()
model.add(LSTM(units = 512, input_shape = (normalised_network_input.shape[1], normalised_network_input.shape[2]), return_sequences=True))
model.add(Dropout(0.3))
model.add(LSTM(512, return_sequences=True))
model.add(Dropout(0.3))
model.add(LSTM(512))
model.add(Dense(256))
model.add(Dropout(0.3))
model.add(Dense(n_vocab, activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam')
```

```python
model.summary()
```

```python
checkpoint = ModelCheckpoint("model.hdf5", monitor='loss', verbose=0, save_best_only=True, mode='min')

model_his = model.fit(normalised_network_input, network_output, epochs=100, batch_size=64, callbacks=[checkpoint])
```

## Predictions

```python
sequence_length = 100
network_input = []


for i in range(len(notes) - sequence_length):
    seq_in = notes[i : i+sequence_length] # Contains 100 values

    network_input.append([ele_to_int[ch] for ch in seq_in])
```

```python
start = np.random.randint(len(network_input) - 1)
```

```python
int_to_ele = dict((num, ele) for num, ele in enumerate(pitchnames))
```

```python
pattern = network_input[start]
prediction_output = []
```

```python
# Generate 200 elements
for note_index in range(200):
    prediction_input = np.reshape(pattern, (1, len(pattern), 1))
    prediction_input = prediction_input/float(n_vocab)
    
    prediction = model.predict(prediction_input, verbose=0)
    idx = np.argmax(prediction)
    
    result = int_to_ele[idx]
    
    prediction_output.append(result)
    
    pattern.append(idx)
    pattern = pattern[1:]
```

```python
prediction_output
```

## Create MIDI File

```python
offset = 0 # Time
output_notes = []


for pattern in prediction:
    
    # if pattern is a chord
    if ('+' in pattern) or pattern.isdigit():
        notes_in_chord = pattern.split('+')
        temp_notes = []
        
        for current_note in notes_in_chord:
            new_note = note.Note(int(current_note))
            new_note.storedInstrument = instrument.Piano()
            temp_notes.append(new_note)
            
        new_chord = chord.Chord(temp_notes)
        new_chord.offset = offset
        output_notes.append(new_chord)
    
    
    
    # if pattern is a note
    else:
        new_note = note.Note(pattern)
        new_note.offset = offset
        new_note.storedInstrument = instrument.Piano()
        output_notes.append(new_note)
        
    offset += 0.5
```

```python
midi_stream = stream.Stream(output_notes)
midi_stream.write('midi', fp = 'test_audio.mid')
```
